# Handle with rate limit exceeded

```python
from langchain_community.vectorstores import FAISS
import backoff  # for exponential backoff
import warnings
warnings.filterwarnings("ignore")

@backoff.on_exception(backoff.expo, openai.RateLimitError)
def completions_with_backoff(**kwargs):
    return FAISS.from_documents(**kwargs)


# completions_with_backoff(model="gpt-3.5-turbo", messages=[{"role": "user", "content": "Once upon a time,"}])
# db = FAISS.from_documents(docs, embedding)

vector_db = completions_with_backoff(documents=docs,embedding=embedding)
vector_db.save_local("faiss_index")
```

```python
new_db = FAISS.load_local("faiss_index", embedding)

docs = new_db.similarity_search('Что дают за совершение убийство?',k=10)
```

```python
pretty_print_docs(docs)
```

# Compressor, для нахождение нужного документа. Модель может запутаться из большего количество данных и выбрать одинаковые или нерелевантные ответы

```python
from langchain.retrievers import ContextualCompressionRetriever
from langchain.retrievers.document_compressors import LLMChainExtractor
from langchain.llms import OpenAI
embedding = OpenAIEmbeddings(api_key='')

llm = OpenAI(temperature=0, api_key='') 
compressor = LLMChainExtractor.from_llm(llm)
compression_retriever = ContextualCompressionRetriever(
    base_compressor=compressor, base_retriever=new_db.as_retriever(search_type = "mmr",search_kwargs={"k":10} )
)
```

```python
retriever = new_db.as_retriever(search_type = "mmr",search_kwargs={"k":10} )
```

```python
ans = retriever.get_relevant_documents("Что дают за совершение убийство?")
```

```python
pretty_print_docs(ans)
```

```python
compressed_docs = compression_retriever.get_relevant_documents(
    "Что дают за убийство?", 
)
pretty_print_docs(compressed_docs)
```
